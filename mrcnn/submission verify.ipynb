## Verify submission csv

```python
import csv
```

```python
submission = []
with open('submission.csv') as f:
    rows = csv.DictReader(f)
    for r in rows:
        submission.append(r)
```

```python
for row in submission:
    if float(row['PixelCount']) < 1:
        print('not good')
```

```python
# we could lower the confidence threshold
for row in submission:
    if float(row['Confidence']) < 0.80:
        print('not good')
```

```python
for i, row in enumerate(submission):
    if i % 100 == 0:
        print(i)
    encoding = row['EncodedPixels']
    total = 0
    res = encoding.split('|')
    res = res[:-1]
    for chunk in res:
        start, length = chunk.split(' ')
        start = int(start)
        length = int(length)
        if start >= (3384 * 2710) - 1 or start <= 0:
            print('start out of bounds', start)
        
        end = start + length
        if end >= (3384 * 2710) - 1 or end <= 0:
            print('end out of bounds', end)
        
        total += length
    
    if total != int(row['PixelCount']):
        print('length not equal', length, int(row['PixelCount']))
        
```
